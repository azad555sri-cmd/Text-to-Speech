```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''welcome to data science, ai, genai, agentic ai program. thank you. ''', lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')      # (if wav is not working change extension to mp3)
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```
